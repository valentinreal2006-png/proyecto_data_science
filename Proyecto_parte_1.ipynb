```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/data sciences/data_choques_listo_CSV.csv', encoding='latin-1', sep=';')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
df['tipo_de_carretera'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Cantidad de choques por tipo de carretera')
plt.xlabel('Tipo de carretera')
plt.ylabel('Cantidad')
plt.show()

print(df['tipo_de_carretera'].value_counts(normalize=True)*100)
```

# Interpretación:

#La mayoría de los accidentes ocurren en carreteras de tipo autopistas, lo cual podría relacionarse con mayor flujo vehicular o velocidad promedio.

```python
plt.figure(figsize=(6,4))
df['clima'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribución de choques según el clima')
plt.xlabel('Condición climática')
plt.ylabel('Cantidad')
plt.show()
```

#Interpretación:

#Se observa que los accidentes ocurren con mayor frecuencia bajo condiciones de clima despejado, lo que sugiere que el clima adverso podría influir en la probabilidad de choque.
